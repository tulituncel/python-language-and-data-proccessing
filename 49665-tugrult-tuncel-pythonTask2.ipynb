```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def read_csv(file_path):
    """
    Read CSV file and return DataFrame.
    """
    df = pd.read_csv(file_path, sep=';')
    df['ita_is_calibration_data'] = df['ita_is_calibration_data'].astype(bool)
    return df

def filter_calibration_data(df):
    """
    Create a new DataFrame with non-calibration data.
    """
    non_calibration_data = df[df['ita_is_calibration_data'] == False]
    return non_calibration_data

def plot_motion_data(df, title):
    """
    Plot motion data for the entire dataset.
    """
    plt.scatter(df['ita_mili'], df['ita_x'], label='X-axis motion')
    plt.scatter(df['ita_mili'], df['ita_y'], label='Y-axis motion')
    plt.scatter(df['ita_mili'], df['ita_z'], label='Z-axis motion')
    plt.title(title)
    plt.legend()

def plot_random_subset(df, n, title):
    """
    Plot a subset of motion data for n randomly selected consecutive rows.
    """
    random_subset = df.sample(n=n)
    plt.scatter(random_subset['ita_mili'], random_subset['ita_x'], label='X-axis motion')
    plt.scatter(random_subset['ita_mili'], random_subset['ita_y'], label='Y-axis motion')
    plt.scatter(random_subset['ita_mili'], random_subset['ita_z'], label='Z-axis motion')
    plt.title(title)
    plt.legend()

def main():
    # CSV file should be in the same folder
    file_path = '/content/sample_data/freeAcc01.csv'

    # Read CSV and preprocess data
    df = read_csv(file_path)
    non_calibration_data = filter_calibration_data(df)

    # Plot motion data for the entire non-calibration dataset
    plt.figure(figsize=(12, 6))
    plt.subplot(2, 1, 1)
    plot_motion_data(non_calibration_data, 'Motion data for the entire non-calibration dataset')

    # Plot a subset of motion data for 20 randomly selected consecutive rows
    plt.subplot(2, 1, 2)
    plot_random_subset(non_calibration_data, n=20, title='Motion data for a randomly selected subset of 20 consecutive rows')

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```
